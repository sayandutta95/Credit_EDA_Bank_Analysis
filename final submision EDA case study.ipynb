## checking the percentage of null values of app1 dataset --

```python
cols = app1.isnull().mean().reset_index()
cols[0] = cols[0]*100
cols = cols.rename(columns={"index":"Column Name",0:"%Missing Values"})
cols
```

```python
# sorting the columns of null values in descending order.
cols= cols.sort_values(by="%Missing Values",ascending=False)
cols
```

```python
#according to the qstn the thresshold value of the null value counting is 40%(as calculated),
#so we got 49 such columns having null values more than 40% & we had decided to drop those columns--

cols_to_drop = list(cols[cols["%Missing Values"]>40]["Column Name"])
print(cols_to_drop)
print(len(cols_to_drop))
```

```python
##dropping the cols_to_drop which contains all the columns having more than 40% null values.
app2 = app1.drop(cols_to_drop,axis=1)  
app2.shape
```

```python
##visualization of dataset after dropping the columns.
app2.head()  
```

```python
#finding the unique values of the columns of app2 and making dataset--
app2_uni = app2.nunique().reset_index()
app2_uni = app2_uni.rename(columns={"index":"col name",0:"unique val"})
app2_uni
```

```python
#dividing the columns in categorical(cat) and continuos(cont) form based on the unique values of the columns--
cat = list(app2_uni[app2_uni["unique val"]<200]["col name"])
cont = list(app2_uni[app2_uni["unique val"]>200]["col name"])
```

## Univariate analysis of the cat and cont columns --

```python
for i in cat:
    print("Countplot of",i)
    sns.countplot(app2[i])
    plt.xticks(rotation=90,fontsize=10)
    plt.show()
```

```python
for i in cont:
    print("Boxplot of",i)
    sns.boxplot(app2[i])
    plt.show()
```

- looking to the above univariate analysis of each of the columns we can impute the null values using median for the cont columns and the cat columns by the mode values.

```python
for i in cont:
    if app2[i].isnull().mean()>0:
        imp_val=app2[i].median()
        app2[i]=app2[i].fillna(imp_val)
```

```python
for i in cat:
    if app2[i].isnull().mean()>0:
        imp_val=app2[i].mode()[0]
        app2[i]=app2[i].fillna(imp_val)
```
